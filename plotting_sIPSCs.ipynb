```python
import numpy as np
import matplotlib.pyplot as plt
import pyabf
import os

def plot_single_trace(abf_path):
    
    # Load ABF file
    abf = pyabf.ABF(abf_path)
    
    # Get data
    time = abf.sweepX
    current = abf.sweepY
    
    # Create plot
    plt.figure(figsize=(12, 6))
    plt.plot(time, current, 'k-', linewidth=1)
    plt.ylim(-60, -20)  # Adjust y-limits as needed
    plt.xlim(0, 6)
    # Labels
    plt.xlabel('Time (s)')
    plt.ylabel('current (pA)')
    plt.title(f'sIPSC Recording - {os.path.basename(abf_path)}')
    
    ax = plt.gca()
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    
    plt.tight_layout()
    plt.show()

# Usage
abf_path ="Z:\\Group Members\\Rima\\Ephys_NE\\DATA\\ntsr1_sIPSCs_sEPSCs\\July 2025 Ntsr sEPSCs sIPSCs\\M5 C2 sIPSCs 002 base.abf"
plot_single_trace(abf_path)
```
